```python
import pandas as pd
data=pd.read_excel('d.xlsx',encoding='utf-8')
df=pd.DataFrame(data)
import jieba
content=df.content.values.tolist()#转换为列表
content_s=[]
for line in content:
    current_segment=jieba.lcut(line)
    if len(current_segment)>1 and current_segment!='\r\n':#换行符
        content_s.append(current_segment)
df_content=pd.DataFrame({'content_s':content_s})#将列表转换成字典
stopwords=pd.read_csv('stopwords.txt',index_col=False,quoting=3,names=['stopword'])
def drop_stopwords(contents,stopwords):
    contents_clean=[]
    all_words=[]
    for line in contents:
        line_clean=[]
        for word in line:
            if word not in stopwords:
                continue
            line_clean.append(word)
            all_words.append(str(word))#准备做词云
        contents_clean.append(line_clean)
        return contents_clean,all_words
contents=df_content.content_s.values.tolist()
stopwords=stopwords.stopword.values.tolist()
contents_clean,all_words=drop_stopwords(contents,stopwords)
df_content=pd.DataFrame({'contents_clean':contents_clean})
df_content.head(5)
df_all_words=pd.DataFrame({'all_words':all_words})
import numpy
words_count=df_all_words.groupby(by=['all_words'])['all_words'].agg({'count':numpy.size})
words_count=words_count.reset_index().sort_values(by=['count'],ascending=False)
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize']=(10.0,5.0)
wordcloud=WordCloud(font_path='/data/simhei.ttf',background_color='white')
word_frequence={x[0]:x[1] for x in words_count.head(100).values}
wordcloud=wordcloud.fit_words(word_frequence)
plt.imshow(wordcloud)
```

```python
#运用结巴分词
import jieba.analyse
index = 100
print(df['content'][index])
content_S_str=''.join(content_s[index])
print(' '.join(jieba.analyse.extract_tags(content_S_str,topK=5,withWeight=False)))
```

```python
from gensim import corpora,models,similarities
import gensim
#做映射，相当于词袋
dictionary=corpora.Dictionary(contents_clean)
corpus=[dictionary.doc2bow(setence) for setence in contents_clean]
lda=gensim.models.ldamodel.LdaModel(corpus=corpus,id2word=dictionary,num_topics=20)#类似Kmeans自己指定K的值
#一号分类结果
print(lda.print_topic(1,topn=5))
for topic in lda.print_topics(num_topics=20,num_words=5):
    print(topic[1])
```
