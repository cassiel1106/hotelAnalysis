```python
import pandas as pd
data=pd.read_excel('民宿酒店评论.xlsx',names=['comments'],encoding='utf-8')
df_comments=pd.DataFrame(data)
import jieba
comments=df_comments.comments.values.tolist()
comment_s=[]
for line in comments:
    current_segment=jieba.lcut(line)   #进行分词
    if len(current_segment)>1 and current_segment!='\r\n':
        comment_s.append(current_segment)
df_comment=pd.DataFrame({'comment_s':comment_s})
commentlist=df_comment.comment_s.values.tolist()
df_comment1=[]
all_comments=[]
for l in comments:   #滤掉非中文字符
    l_clean=[]
    for char in l:
        if char >= u'\u4e00' and char <= u'\u9fa5':
            
            l_clean.append(char)
            all_comments.append(str(char))
    df_comment1.append(l_clean)
df_all_comment=pd.DataFrame({'all_comments':all_comments})
import numpy
words_count=df_all_comment.groupby(by=['all_comments'])['all_comments'].agg({'count':numpy.size})#词频统计
words_count=words_count.reset_index().sort_values(by=['count'],ascending=False)
from wordcloud import WordCloud  #制作词云
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize']=(10.0,5.0)
wordcloud=WordCloud(font_path='/data/simhei.ttf',background_color='white')
word_frequence={x[0]:x[1] for x in words_count.values}
wordcloud=wordcloud.fit_words(word_frequence)
plt.imshow(wordcloud)
```

```python
import pandas as pd
data=pd.read_excel('民宿酒店评论.xlsx',names=['comments'],encoding='utf-8')
df_comments=pd.DataFrame(data)
import jieba
comments=df_comments.comments.values.tolist()
comment_s=[]
for line in comments:
    current_segment=jieba.lcut(line)
    if len(current_segment)>1 and current_segment!='\r\n':
        comment_s.append(current_segment)
df_comment=pd.DataFrame({'comment_s':comment_s})
commentlist=df_comment.comment_s.values.tolist()
stopwords=pd.read_csv('Jstopwords.txt',index_col=False,quoting=3,names=['stopword'],encoding='gbk')
def drop_stopwords(comments,stopwords):
    comments_clean=[]
    all_comments=[]
    for line in comments:
        line_clean=[]
        for word in line:
            if word in stopwords:
                continue
            line_clean.append(word)
            all_comments.append(str(word))
        comments_clean.append(line_clean)
        return comments_clean,all_comments
comments=df_comment.comment_s.values.tolist()
#print(comments)
stopwords=stopwords.stopword.values.tolist()
#print(stopwords)
comments_clean,all_comments=drop_stopwords(comments,stopwords)
#df_clean_comment=pd.DataFrame({'comments_clean':comments_clean})
#分组计算词频
df_all_comment=pd.DataFrame({'all_comments':all_comments})
import numpy
words_count=df_all_comment.groupby(by=['all_comments'])['all_comments'].agg({'count':numpy.size})
words_count=words_count.reset_index().sort_values(by=['count'],ascending=False)
#画一个词云
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize']=(10.0,5.0)
wordcloud=WordCloud(font_path='/data/simhei.ttf',background_color='white')
word_frequence={x[0]:x[1] for x in words_count.values}
wordcloud=wordcloud.fit_words(word_frequence)
plt.imshow(wordcloud)
```
